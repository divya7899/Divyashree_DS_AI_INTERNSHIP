Machine learning is the one of the subset of AI.It is model which will learn from past experience and perform the specific task and through that the performance will be measured.Machine learning mainly helps us to work on the numerical data.Which handles the numerical data and handle the data and preprocess the data and learn from the training data and generated model will helps us to analyse the data and helps to predict the result.Machine learning has mainly 4 types that are: Supervised learning ,unsupervised learning, Semisupervised learning ,Reinforcement learning. For supervised learning real world example:Email spam detection,for unsupervised learning :Clustering students based on performance, for reinforcement learning :self driving car .Machine learning which is training machine for performing specific task so that human work ,mistakes ,time will be reduced.ML model which is model which mimic human task and learns from past data.
