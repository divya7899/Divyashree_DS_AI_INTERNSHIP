Correctly classify real-world problems into the appropriate Machine Learning category.

Choosing the correct learning framework is critical. If the wrong type is selected, the model will fail even before implementation.

#Supervised learning
Supervised Learning is a type of machine learning where the dataset contains labeled data (correct answers).  
The model learns a mapping from inputs to outputs.


Real-World Examples:
 Spam Detection → Spam / Not Spam
 House Price Prediction → Predict price value

#Unsupervised Learning

Unsupervised Learning is used when the dataset has no labels.  
The model discovers hidden patterns or clusters within the data.

Real-World Examples
Customer Segmentation
Market Basket Analysis
Grouping similar products
Detecting unusual transaction patterns

#Reinforcement Learning
Reinforcement Learning is a type of learning where an agent interacts with an environment and learns through rewards and penalties.


Real-World Examples
Self-driving cars
Game-playing AI
Robotics movement control
Dynamic pricing optimization


#Comparison Table

| ML Type | Data Type | Learning Method | Real-World Examples |
|----------|------------|----------------|----------------------|
| Supervised Learning | Labeled Data | Learn input → output mapping | Spam Detection, House Price Prediction |
| Unsupervised Learning | Unlabeled Data | Find hidden clusters/patterns | Customer Segmentation, Market Basket Analysis |
| Reinforcement Learning | Reward-based | Learn via rewards & penalties | Self-driving Cars, Game AI |
